```python
df= df.copy()
```

```python
# assinging features and target
x= df.drop("R", axis=1) 
y = df['R']
```

```python
# normalizing and standarizing  values
scaler = StandardScaler()
scaler
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
# splitting date for training 
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state=0)
```

```python
# creating model for calssification
sonar_log_reg = LogisticRegression(max_iter=1000,C= 0.1, penalty= 'l2', solver = 'saga')
sonar_log_reg
```

```python
from sklearn.model_selection import GridSearchCV


# Defining features and parameters
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}

#  Grid Search
grid = GridSearchCV(estimator=sonar_log_reg, param_grid=param_grid, cv=5, scoring='accuracy')
grid.fit(x_train, y_train)

# Results
print(f"Best Parameters: {grid.best_params_}")
print(f"Best Cross-Validation Score: {grid.best_score_}")
```

```python
# fitting model 
sonar_log_reg.fit(x_train, y_train)
```

```python
# calculating prediction
y_pred = sonar_log_reg.predict(x_test)
```

```python
y_pred
```

```python
# checking accuracy
acc = accuracy_score(y_test, y_pred)
```
